```python
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/pycaret/pycaret-demo-dphi/main/insurance.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
from pycaret.regression import *
s = setup(data, target = 'charges')
```

```python
models()
```

```python
best_model=compare_models(exclude=['ransac','tr','huber','br'])
```

```python
best_model
```

```python
predict_model(best_model,data=data)
```
