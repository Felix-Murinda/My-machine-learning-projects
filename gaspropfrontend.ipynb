```python
from tkinter import *
window = Tk()
window.title("GAS PROPERTIES")
import gaspropbackend

def rho_command():
    R=10.7
    Ma = float(e1.get())
    P = float(e2.get())
    T = float(e3.get())
    rho = gaspropbackend.density(Ma,P,R,T)
    list2.insert(rho)
    e4.delete(0, END)
    e4.insert(END, "{:.2f}".format(Rs))
    e5.delete(0, END)
    e5.insert(END, "{:.2f}".format(Rs))
    
    
def Vs_command():
    R=10.7
    Ma = float(e1.get())
    P = float(e2.get())
    T = float(e3.get())
    rho = gaspropbackend.density(Ma,P,R,T)
    Vs = gaspropbackend.specificvolume(rho)
    list3.insert(Vs)
    
def Tr_command():
    Tc= float(e4.get())
    T = float(e3.get())  
    Tr = gaspropbackend.reducedtemperature(T,Tc)
    list1.insert(Tr)
    
def Pr_command():
    Pc= float(e5.get())
    P = float(e2.get())
    Pr = gaspropbackend.reducedpressure(P,Pc)
    list5.insert(Pr)
    
def Bg_command():
    Z= float(e6.get())
    P = float(e2.get())
    T = float(e3.get())

    Bg = gaspropbackend.volumefactor(Z,T,P)
    list7.insert(Bg)
    
def Eg_command():
    Bg = gaspropbackend.volumefactor(Z,T,P)
    Eg = gaspropbackend.expansionfactor(Bg)
    list3.insert(Eg)

l1=Label(window, text="APPARENT MOLECULAR WEIGHT(Ma)")
l1.grid(row=1, column=0)

molecularweight_text= StringVar()
e1= Entry(window,textvariable= molecularweight_text)
e1.grid(row=1, column=1)


l2=Label(window, text="PRESSURE (P)")
l2.grid(row=2, column=0)

pressure_text= StringVar()
e2= Entry(window,textvariable= pressure_text)
e2.grid(row=2, column=1)

l3=Label(window, text="TEMPERATURE (T)")
l3.grid(row=2, column=3)

temperature_text= StringVar()
e3= Entry(window,textvariable= temperature_text)
e3.grid(row=2, column=4)

b2= Button(window,text="GET DENSITY", height=5, width = 10, command= rho_command)
b2.grid(row=3, column=0)

list2=Listbox(window,height=5, width=10)
list2.grid(row=3, column=1,)

b3= Button(window,text="GET SPECIFIC VOLUME", height=5, width = 20, command= Vs_command)
b3.grid(row=3, column=4)

list3=Listbox(window,height=5, width=10)
list3.grid(row=3, column=9,)

l4=Label(window, text="PSEUDOCRITICAL TEMPERATURE(Tpc)")
l4.grid(row=4, column=0)

pseudotemperature_text= StringVar()
e4= Entry(window,textvariable= pseudotemperature_text)
e4.grid(row=4, column=1)

l5=Label(window, text="PSEUDOCRITICAL PRESSURE(Ppc)")
l5.grid(row=4, column=3)

pseudopressure_text= StringVar()
e5= Entry(window,textvariable= pseudopressure_text)
e5.grid(row=4, column=4)

l13= Label(window,text=" Z FACTOR (Z)")
l13.grid(row=6, column=0)

zfactor_text= StringVar()
e6= Entry(window,textvariable= zfactor_text)
e6.grid(row=6, column=1)

b5= Button(window,text="GET REDUCED PRESSURE(Pr)", height=5, width = 22,command= Pr_command)
b5.grid(row=5, column=0)

list5=Listbox(window,height=5, width=10)
list5.grid(row=5, column=1,)

b6= Button(window,text="GET REDUCED TEMPERATURE(Tr)", height=5, width = 25, command=Tr_command)
b6.grid(row=5, column=4)

list1=Listbox(window,height=5, width=10)
list1.grid(row=5, column=9,)

b7= Button(window,text="GET GAS FORMATION FACTOR (Bg)", height=5, width = 27, command=Bg_command)
b7.grid(row=7, column=0)

list7=Listbox(window,height=5, width=10)
list7.grid(row=7, column=1,)

b8= Button(window,text="GET EXPANSION FACTOR (Eg)", height=5, width = 25, command=Eg_command)
b8.grid(row=7, column=4)

list8=Listbox(window,height=5, width=10)
list8.grid(row=7, column=9,)


window.mainloop()
```
